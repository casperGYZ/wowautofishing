```python
DECIBEL_THRESHOLD = -40
```

```python
stream = p.open(format=pyaudio.paInt16, channels=1, rate=44100, input=True, input_device_index=4)
#timmer_1 = time.time()-601
time.sleep(2)
while True:
    #pyautogui.press('0')
    #time.sleep(round(random.uniform(1,1.2)))
    #pyautogui.press('9')    
    #time.sleep(round(random.uniform(0.3,0.5)))
    #pyautogui.press('9') 
    data = stream.read(1024, exception_on_overflow = False)
    time.sleep(round(random.uniform(0.1,0.3)))
    pyautogui.press(']')
    data = stream.read(1024, exception_on_overflow = False)
    time.sleep(2)
    data = stream.read(1024, exception_on_overflow = False)
    start_time = time.time()
    data = stream.read(1024, exception_on_overflow = False)
    while True:
        data = stream.read(1024, exception_on_overflow = False)
        data = np.frombuffer(data, dtype=np.int16) / 32768.0
        rms = np.sqrt(np.mean(data**2))
        db = np.round(20 * np.log10(rms / ((2**15) / 32768.0)))
        if db > DECIBEL_THRESHOLD:
            print(db)
            time.sleep(round(random.uniform(0.5,1.2)))
            pyautogui.press('[')
            time.sleep(round(random.uniform(1,1.5)))
            break
        if time.time() - start_time >= 22:
            break
```
